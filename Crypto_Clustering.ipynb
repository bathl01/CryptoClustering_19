```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas as hvplot
import holoviews as hv
import os
# Set OMP_NUM_THREADS environment variable to 1
os.environ['OMP_NUM_THREADS'] = '1'
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MiniBatchKMeans
print(os.environ['OMP_NUM_THREADS'])

```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
crypto_scaled = StandardScaler().fit_transform(df_market_data)
crypto_scaled
```

```python
# Create a DataFrame with the scaled data
crypto_df = pd.DataFrame(
    crypto_scaled,
    

# Copy the crypto names from the original data
columns=df_market_data.columns,

# Set the coinid column as index
index=df_market_data.index)


# Display sample data
crypto_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_orig = list(range(1, 12))
k_orig
```

```python


# Create an empty list to store the inertia values
inertia_orig = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k_orig:
    orig_model = MiniBatchKMeans(n_clusters=i, random_state=1, batch_size=max(3072, len(crypto_df)), max_iter=100, n_init=3)   
    orig_model.fit(crypto_df)
    inertia_orig.append(orig_model.inertia_)
      
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_orig = {"k_orig": k_orig, "inertia_orig": inertia_orig}

# Create a DataFrame with the data to plot the Elbow curve
elbow_orig_df = pd.DataFrame(elbow_orig)
elbow_orig_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
orig_elbow_plot = elbow_orig_df.hvplot.line(
    x="k_orig",
    y="inertia_orig",
    title= "Original Elbow Curve",
    xticks=k_orig
)
orig_elbow_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** the best value for 'K' is 4 with the original data

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model_4 = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
model_4.fit(crypto_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_4 = model_4.predict(crypto_df)

# Print the resulting array of cluster values.
k_4
```

```python
# Create a copy of the DataFrame
crypto_4_df = crypto_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
crypto_4_df['predicted_cluster'] = k_4

# Display sample data
crypto_4_df
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# Define a colormap for the clusters
# colors to use
orig_color = ['red', 'orange', 'brown', 'pink']
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
# Create the scatter plot with specified colors
orig_scatter_plot = crypto_4_df.hvplot.scatter(
    x='price_change_percentage_24h',
    y='price_change_percentage_7d',
    by='predicted_cluster',
    title='Original Scatter Plot',
    hover_cols=['coin_id', 'predicted_cluster'],  # Add 'coin_id' to hover_cols
    color=orig_color  
)
 
# Display the scatter plot
orig_scatter_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
crypto_pca = pca.fit_transform(crypto_4_df)
# View the first five rows of the DataFrame. 
crypto_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

```python
#calculate the Total Explained Variance by summing all 3 Explained Variance Ratios
sum(pca.explained_variance_ratio_)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:**  Total Explained Variance = the sum of the 3 Explained Variance Ratios

  0.34871677, 0.31363391, 0.22627118 = 0.8886218549859446

```python
# Create a new DataFrame with the PCA data.
crypto_pca_df = pd.DataFrame(
    crypto_pca,
    columns = ["PCA1", "PCA2", "PCA3"]
)
crypto_pca_df

# Copy the crypto names from the original data
crypto_pca_df['coin_id'] = list(crypto_4_df.index)
crypto_pca_df

# Set the coinid column as index
crypto_pca_df = crypto_pca_df.set_index('coin_id')

# Display sample data
crypto_pca_df
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_pca= list(range(1,12))
k_pca
```

```python
# Create an empty list to store the inertia values
inertia_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k_pca:
    pca_model = MiniBatchKMeans(n_clusters=i, random_state=1, batch_size=max(3072, len(crypto_pca_df)), max_iter=100, n_init=3)   
    pca_model.fit(crypto_pca_df)
    inertia_pca.append(pca_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_pca = {"k_pca": k_pca, "inertia_pca": inertia_pca}
# Create a DataFrame with the data to plot the Elbow curve
elbow_pca_df = pd.DataFrame(elbow_pca)
elbow_pca_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
pca_elbow_plot = elbow_pca_df.hvplot.line(
    x="k_pca",
    y="inertia_pca",
    title= "PCA Elbow Curve",
    xticks=k_pca
)
pca_elbow_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** the best value for 'k' is 4 with the PCA data


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, it is the same 'k' value as found using the original data
